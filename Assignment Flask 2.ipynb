```python
# Q3. Why is redirect() used in Flask?

# redirect() is used in the flask to redirect to the different url. It redirects the user to the different url after they have performed the certain activity.

from flask import Flask,redirect

app = Flask(__name__)

@app.route("/helloworld")
def home():
    return "<h1>Hello world!</h1>"

@app.route("/redirect")
def redirect_method():
    return redirect("/helloworld")

if __name__=="__main__":
    app.run(host = "0.0.0.0")
    

# in this program app.route("/redirect") will redirect the user "/helloworld" url. 
```
