```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

# plotting the data on x and y axis and presenting the data as dots on the graph

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Note: Also add title, xlabel, and ylabel to the plot.

plt.scatter(x,y)
plt.title("My first scatter plt")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:

# to draw multiple plot together. 
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# plot 1
plt.subplot(2,2,1)
plt.plot(x1,y1)

# plot 2 
plt.subplot(2,2,2)
plt.plot(x2,y2)

# plot 3
plt.subplot(2,2,3)
plt.plot(x3,x4)

# plot 4
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

# graphical representation of categories as rectangulaar bars which is proportional to its values. 
# bar plot is graphical representation of categorical vs numerical data

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.title("Bar plot")
plt.xlabel("Names of the companies")
plt.ylabel("Profit according to the companies")
plt.show()
```

```python
plt.barh(company,profit)
plt.title("Horizontal bar plt")
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

# is used to display the distribution of numerical data, it shows first & third quartile, min and max, also the middle line which shows median. It also shows if their is any outliers in the data

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.figure(figsize = (10,7))
plt.boxplot(data)
plt.show()
```
