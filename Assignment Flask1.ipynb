```python
# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

from flask import Flask, request

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1> Hello World!</h1>"

if __name__ == "__main__":
    app.run(host = "0.0.0.0")
```

```python
# Q3. What is App routing in Flask? Why do we use app routes?

# it is a method for mapping urls to python functions in a flask applicaton. 
# The reason we use app routes to tell the program where to go and execute that program. 

from flask import Flask, request

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1> Hello World!</h1>"

if __name__ == "__main__":
    app.run(host = "0.0.0.0")
    
# For example in this program we are telling the program to the route which has "/". 
```

```python
# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
# route to show the following details:
# Company Name: ABC Corporation
# Location: India
# Contact Detail: 999-999-9999

from flask import Flask,request

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1> Welcome to ABC Corporation</h1>"

@app.route("/")
def details():
    return " Company name: ABC Corporation <br> Location: India <br> Contact Detail: 999-999-999 "


if __name__ == "__main__":
    app.run(host = "0.0.0.0")

```
